## Ploting the PCAs for question 1

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

## Load file eigenvec file as dataframe

```python
df= pd.read_csv("/Users/cmdb/qbb2021-answers/week3/plink.eigenvec", sep=" ",header=None)
```

```python
df.columns=['FID','IID','PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16','PC17','PC18','PC19','PC20']
df
##rename the columns
```

```python
x=df.loc[:,'PC1'].values
y=df.loc[:,'PC2'].values

#pull out the first two PCA components
```

## make a pca scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title('PCA plot')
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
plt.show()
    
    
```

```python
allele_freq=[]
for line in open("freq.frq"):
    column=line.strip("\t").split()
    if column[0]!='CHR':
        allele_freq.append(float(column[4]))
#print(allele_freq)
fig, ax=plt.subplots()
ax.hist(allele_freq, bins=80)
ax.set_ylabel("Frequency")
ax.set_title("Allele Freq Count")
plt.show()
```
